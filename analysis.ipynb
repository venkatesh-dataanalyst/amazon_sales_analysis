```python
df=pd.read_csv("amazon.csv")
```

```python
df['discount_percentage']=df['discount_percentage'].str.replace('%',"")
```

```python
df['discount_percentage']=pd.to_numeric(df['discount_percentage'])
```

```python
df['discount_percentage'].head()
```

```python
df['rating']=pd.to_numeric(df['rating'],errors='coerce')
```

```python
df[['discount_percentage','rating']].corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='discount_percentage',y='rating',data=df)
plt.title("Discount vs Rating Relationship")
plt.show()
```
